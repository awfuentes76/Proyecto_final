## 3. Clustering y segmentación

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd

# Seleccionar las columnas relevantes para clustering
variables_clustering = [
    'RMH', 'LENGUA', 'JUNTS_cambio', 
    'ERC_cambio', 'PSC_cambio', 'CUP_cambio', 
    'ECP_cambio'
]

# Normalizar los datos
scaler = StandardScaler()
df_clustering = pd.DataFrame(scaler.fit_transform(df[variables_clustering]), columns=variables_clustering)

# Verificar la normalización
print(df_clustering.describe())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Determinación del número óptimo de clusters usando el método del codo (Elbow Method)
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_clustering)  
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('SSE (Suma de Cuadrados de las Distancias a los Centroides)')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.show()

# Aplicar K-means con el número óptimo de clusters determinado por el método del codo
optimal_k = 4  # Ajusta este valor según el gráfico del codo
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(df_clustering)  

# Visualizar los clusters en el espacio de las dos primeras componentes principales
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_clustering['RMH'], y=df_clustering['LENGUA'], hue=df['cluster'], palette='viridis')
plt.title('Clusters según RMH y LENGUA')
plt.show()
```

```python
# Analizar cambios en el voto por cluster
cambios_por_cluster = df.groupby('cluster')[['JUNTS_cambio', 'ERC_cambio', 'PSC_cambio', 'CUP_cambio', 'ECP_cambio']].describe()

# Mostrar los resultados descriptivos
print(cambios_por_cluster)


# Crear una figura con subplots para los boxplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 filas, 3 columnas

# Ajustar espacio entre los subplots
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Lista de cambios en el voto para iterar
cambios_voto = ['JUNTS_cambio', 'ERC_cambio', 'PSC_cambio', 'CUP_cambio', 'ECP_cambio']

# Títulos para cada gráfico
titulos = [
    'Cambio en el voto de JUNTS por cluster',
    'Cambio en el voto de ERC por cluster',
    'Cambio en el voto de PSC por cluster',
    'Cambio en el voto de CUP por cluster',
    'Cambio en el voto de ECP por cluster'
]

# Iterar sobre las variables y los ejes para crear los boxplots
for i, ax in enumerate(axes.flat[:5]):  # [:5] para usar solo 5 subplots
    sns.boxplot(x='cluster', y=cambios_voto[i], data=df, ax=ax)
    ax.set_title(titulos[i])
    ax.set_xlabel('Cluster')
    ax.set_ylabel('Cambio de Voto')

# Eliminar el subplot vacío si es necesario
fig.delaxes(axes[1, 2])

# Mostrar el gráfico
plt.show()
```

Heterogeneidad dentro de los clusters: Los resultados de la clusterización sugieren que existen diferencias significativas entre los grupos de secciones censales, basadas en características socioeconómicas, lingüísticas y electorales. Sin embargo, dentro de cada cluster, es probable que todavía existan variaciones en el comportamiento electoral que no han sido completamente capturadas por la clusterización. Un análisis de regresión multinivel permitirá modelar esta variabilidad dentro y entre clusters.

Estructura jerárquica de los datos: Los datos que tienes tienen una estructura jerárquica natural: secciones censales anidadas dentro de clusters. La regresión multinivel es adecuada para este tipo de datos porque permite modelar tanto los efectos de los predictores a nivel de sección (nivel 1) como a nivel de cluster (nivel 2).

2. Relación con la Clusterización:

Modelos específicos por cluster: La clusterización ya ha identificado grupos de secciones con patrones similares. Ahora, la regresión multinivel permitirá investigar si los efectos de variables como RMH, LENGUA o los cambios en el voto varían sistemáticamente entre clusters. En otras palabras, podremos observar si las relaciones entre variables difieren entre clusters, lo que podría proporcionar información valiosa sobre cómo y por qué los votantes en diferentes áreas responden de manera diferente a las mismas condiciones socioeconómicas o políticas.
Interceptos y pendientes aleatorios: La regresión multinivel te permitirá especificar interceptos y pendientes aleatorios para los clusters, lo que te permitirá modelar cómo las relaciones entre las variables dependen del contexto del cluster. Por ejemplo, podrías descubrir que el impacto de la renta media por hogar (RMH) en el cambio en el voto para un partido es diferente en clusters con diferentes composiciones lingüísticas.

Siguiente paso:
Definir los modelos multinivel: Podrías comenzar por definir modelos con RMH, LENGUA, y otros predictores clave a nivel de sección (nivel 1) y considerar el cluster como un factor de nivel 2. Esto te permitirá estimar si y cómo las relaciones entre estas variables cambian según el cluster.
Incorporar variables de nivel 2: Además de los predictores de nivel 1, podrías considerar incluir predictores a nivel de cluster que puedan influir en los patrones observados, como el cluster promedio de participación o un indicador de diversidad lingüística dentro del cluster.
En resumen, el análisis de regresión multinivel permitirá capturar las complejidades de los datos y entender mejor las dinámicas electorales en función de los factores socioeconómicos y lingüísticos identificados en los clusters. Este enfoque nos proporcionará un marco más robusto para interpretar los cambios en el voto y los factores que los impulsan.

```python
import statsmodels.formula.api as smf

# Asignamos la columna de cluster al dataset original si no lo has hecho ya
df['cluster'] = kmeans.fit_predict(df_clustering)

# Definir el modelo multinivel para el cambio en el voto de JUNTS
modelo_junts = smf.mixedlm("JUNTS_cambio ~ ACTIVEM23 + CUP23", df, groups=df["cluster"]).fit()

# Resumen del modelo
print(modelo_junts.summary())

# Definir el modelo multinivel para el cambio en el voto de ERC
modelo_erc = smf.mixedlm("ERC_cambio ~ CUP23", df, groups=df["cluster"]).fit()

# Resumen del modelo
print(modelo_erc.summary())

# Definir el modelo multinivel para el cambio en el voto de CUP
modelo_cup = smf.mixedlm("CUP_cambio ~ RMH + LENGUA + PES + JUNTS23 + ERC23", df, groups=df["cluster"]).fit()

# Resumen del modelo
print(modelo_cup.summary())
```

```python
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr
import pandas as pd

# Activar la conversión automática entre pandas y R data frames
pandas2ri.activate()

# Cargar el paquete lphom en R
lphom = importr('lphom')

# Función para ejecutar el análisis de lphom
def ejecutar_lphom(df, cols_primera, cols_segunda, nombre_matriz_primera, nombre_matriz_segunda):
    # Selecciona las columnas correspondientes a la primera elección
    matriz_primera = df[cols_primera].values

    # Selecciona las columnas correspondientes a la segunda elección
    matriz_segunda = df[cols_segunda].values

    # Guardar las matrices en archivos CSV para su uso en R
    pd.DataFrame(matriz_primera).to_csv(f'{nombre_matriz_primera}.csv', index=False, header=False)
    pd.DataFrame(matriz_segunda).to_csv(f'{nombre_matriz_segunda}.csv', index=False, header=False)

    # Convertir las matrices a data frames de R
    r_votos_primera = pandas2ri.py2rpy(pd.DataFrame(matriz_primera))
    r_votos_segunda = pandas2ri.py2rpy(pd.DataFrame(matriz_segunda))

    # Ejecutar la función lphom
    resultado = lphom.lphom(r_votos_primera, r_votos_segunda)

    # Mostrar los resultados
    print(f"Resultados para {nombre_matriz_primera} -> {nombre_matriz_segunda}:")
    print(resultado)
    print("\n")

# Primer modelo: Sin participación/abstención
ejecutar_lphom(
    df,
    cols_primera=['JUNTS19', 'ERC19', 'PSC19', 'CUP19', 'ECP19'],
    cols_segunda=['JUNTS23', 'ERC23', 'PSC23', 'CUP23', 'ECP23'],
    nombre_matriz_primera='matriz_primera_primer',
    nombre_matriz_segunda='matriz_segunda_primer'
)

# Segundo modelo: Con participación/abstención
ejecutar_lphom(
    df,
    cols_primera=['JUNTS19', 'ERC19', 'PSC19', 'CUP19', 'ECP19', 'ABSTENCION19'],
    cols_segunda=['JUNTS23', 'ERC23', 'PSC23', 'CUP23', 'ECP23', 'ABSTENCION23'],
    nombre_matriz_primera='matriz_primera_segundo',
    nombre_matriz_segunda='matriz_segunda_segundo'
)

# Tercer modelo: Incluyendo a Activem, sin participación/abstención
ejecutar_lphom(
    df,
    cols_primera=['JUNTS19', 'ERC19', 'PSC19', 'CUP19', 'ECP19'],
    cols_segunda=['JUNTS23', 'ERC23', 'PSC23', 'CUP23', 'ECP23', 'ACTIVEM23'],
    nombre_matriz_primera='matriz_primera_tercero',
    nombre_matriz_segunda='matriz_segunda_tercero'
)

# Cuarto modelo: Incluyendo a Activem y participación/abstención
ejecutar_lphom(
    df,
    cols_primera=['JUNTS19', 'ERC19', 'PSC19', 'CUP19', 'ECP19', 'ABSTENCION19'],
    cols_segunda=['JUNTS23', 'ERC23', 'PSC23', 'CUP23', 'ECP23', 'ACTIVEM23', 'ABSTENCION23'],
    nombre_matriz_primera='matriz_primera_cuarto',
    nombre_matriz_segunda='matriz_segunda_cuarto'
)

# Quinto modelo: Incluyendo a Activem, ARA-PL, VOX y participación/abstención
ejecutar_lphom(
    df,
    cols_primera=['JUNTS19', 'ERC19', 'PSC19', 'CUP19', 'ECP19', 'ABSTENCION19'],
    cols_segunda=['JUNTS23', 'ERC23', 'PSC23', 'CUP23', 'ECP23', 'ACTIVEM23', 'ARA-PL23', 'VOX23', 'ABSTENCION23'],
    nombre_matriz_primera='matriz_primera_quinto',
    nombre_matriz_segunda='matriz_segunda_quinto'
)
```
